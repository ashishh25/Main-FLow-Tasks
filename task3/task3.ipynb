```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load a CSV file
df = pd.read_csv(r"C:\Users\Ashish\Downloads\BigBasket Products.csv\BigBasket Products.csv")
```

```python
print(df.head(10))
```

```python
# Scatter plot for sale_price vs. market_price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sale_price', y='market_price', data=df, color='#3C486B')
plt.title('Scatter Plot of Sale Price vs. Market Price')
plt.xlabel('Sale Price')
plt.ylabel('Market Price')
plt.show()
```

```python
# Histogram of sale_price
plt.figure(figsize=(10, 8))
sns.histplot(df['sale_price'], bins=20, kde=True, color='#246494')
plt.title('Histogram of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar graph of product counts by category
plt.figure(figsize=(12, 8))
category_counts = df['category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values, color='#FC9D5A')
plt.title('Bar Graph of Product Counts by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
